```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]
x2= np.arange(0,5,0.5)
plt.figure(figsize=(5,3),dpi=150)
plt.plot(x,y,label="2x",linestyle="--",marker="^",markersize=10)
plt.plot(x2[:6],x2[:6]**2,label="x^2",   color="red")
plt.plot(x2[5:],x2[5:]**2,label="x^2",linestyle="--" , color="red")
plt.xticks([0,1,2,3,4])
plt.legend()
plt.savefig("my_first_plot")
plt.show()
```
