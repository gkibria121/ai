```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.arange(1,11)*1.0
y_train = (x_train* 200.0 ) +100.0
```

```python
print(x_train,y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

def get_cost(x_arr, y_arr, w, b):
    sum_cost = 0
    m = len(x_arr)
    for i in range(m):
        y = w * x_arr[i] + b
        sum_cost += (y - y_arr[i]) ** 2
    return (1 / (2 * m)) * sum_cost

# Create meshgrid for w and b values
w_arr = np.arange(0, 401)
b_arr = np.arange(0, 401)
W, B = np.meshgrid(w_arr, b_arr)

# Calculate costs for all combinations of w and b
costs = np.zeros(W.shape)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        costs[i, j] = get_cost(x_train, y_train, W[i, j], B[i, j])

# Create 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surface = ax.plot_surface(W, B, costs, cmap='viridis', alpha=0.8)

# Add labels and title
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Cost')
ax.set_title('Cost Function Surface')

# Add color bar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

plt.tight_layout()
plt.show() 
```
