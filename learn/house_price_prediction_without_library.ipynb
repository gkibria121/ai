```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train =np.array([1.0,2.0])
y_train =np.array([300.0,500.0])
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x_train,y_train,marker="x",c="r")
plt.xlabel("Size (1000sft)")
plt.ylabel("Price (1000s)")
plt.show()
```

```python
def compute_model(x_train ,w,b ):
    m = len(x_train)
    f_y = np.zeros(m)
    
    for i  in range( m  ):
        f_y[i] = (x_train[i]*w) + b
    
    return f_y
```

```python
f_w = 200
f_b = 100
y_test = compute_model(x_train=x_train,w=f_w,b=f_b)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x_train,y_train,marker="x",c="r")
plt.plot(x_train,y_test,c='b')
plt.xlabel("Size (1000sqft)")
plt.ylabel("Price (1k$)")
plt.show()
```

```python
x = 1.2
y = compute_model([x],f_w,f_b)
print("house price is {}k$".format(int(y[0])))
```
